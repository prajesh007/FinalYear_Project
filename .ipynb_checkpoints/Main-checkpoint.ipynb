```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
from tkinter import *
from PIL import ImageTk, Image
from tkinter import filedialog
import os
import matplotlib.image as mpimg
from math import sqrt
import time
```

```python
root = Tk()
root.title('Image Viewer')
# def open():
# global my_image
# global filepath
root.file = filedialog.askopenfilename(
    initialdir="", title="select a file", filetypes=(("png files", "*.png"), ("jpg file", "*.jpg")))
# my_label = Label(root, text=root.file).pack()
# my_image = ImageTk.PhotoImage(Image.open(root.file))
# my_image_label = Label(image=my_image).pack()
filepath = os.path.abspath(root.file)


# my_btn = Button(root, text="open file", command=open).pack()

root.mainloop()

ip_img = imread(filepath)
r, g, b = ip_img[:, :, 0], ip_img[:, :, 1], ip_img[:, :, 2]
gamma = 1.04

rc, gc, bc = 0.2126, 0.7152, 0.0722
ch, cw, cc = ip_img.shape

g_img = rc*r**gamma+gc*g**gamma+bc*b**gamma
gh, gw = g_img.shape

fig = plt.figure(1)
img1, img2 = fig.add_subplot(121), fig.add_subplot(122)
img1.set_title('color \nSize: '+str(ch)+'x'+str(cw)+'x'+str(cc))
img1.imshow(ip_img)
img2.set_title('gray scale \nSize: '+str(gh)+'x'+str(gw))
img2.imshow(g_img, cmap=plt.cm.get_cmap('gray'))

fig.show()
plt.show()

open("myfile.jpg", "w")

cv2.imwrite("myfile.jpg", g_img)
cv2.waitKey(0)

src_img=cv2.imread('myfile.jpg')









cv2.destroyAllWindows()
```

# Gaussian Filter

```python
def imgGauss(image2):
    new_image_gauss = cv2.GaussianBlur(image2, (7, 7),0)
    plt.figure(figsize=(11,6))
    plt.subplot(121), plt.imshow(image2, cmap='gray'),plt.title('Original')
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(new_image_gauss, cmap='gray'),plt.title('Gaussian Filter')
    plt.xticks([]), plt.yticks([])
    plt.show()
    
imgGauss(src_img)
cv2.waitKey(0)
```

# Laplacian Filter

```python
def imgLaplac(image2):
    new_image = cv2.Laplacian(image2,cv2.CV_64F)
    plt.figure(figsize=(11,6))
    plt.subplot(121), plt.imshow(image2, cmap='gray'),plt.title('Original')
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(new_image, cmap='gray'),plt.title('Laplacian')
    plt.xticks([]), plt.yticks([])
    # plt.subplot(133), plt.imshow(image2 + new_image, cmap='gray'),plt.title('Resulting image')
    # plt.xticks([]), plt.yticks([])
    plt.show()
    
imgLaplac(src_img)
cv2.waitKey(0)
```

# Sobel Filter

```python
def imgSobel(image):
    gradients_sobelx = cv2.Sobel(image, -1, 1, 0)
    gradients_sobely = cv2.Sobel(image, -1, 0, 1)
    gradients_sobelxy = cv2.addWeighted(gradients_sobelx, 0.5, gradients_sobely, 0.5, 0)


    plt.subplot(2,2,1), plt.imshow(image, cmap='gray')
    plt.title('Original'),plt.xticks([]),plt.yticks([])
    plt.subplot(2,2,2),plt.imshow(gradients_sobelx, cmap='gray')
    plt.title('Sobel x'),plt.xticks([]),plt.yticks([])
    plt.subplot(2,2,3),plt.imshow(gradients_sobely, cmap='gray')
    plt.title('Sobel y'),plt.xticks([]),plt.yticks([])
    plt.subplot(2,2,4),plt.imshow(gradients_sobelxy, cmap='gray')
    plt.title('Sobel x + y'),plt.xticks([]),plt.yticks([])

    plt.show()
    
    
imgSobel(src_img)
cv2.waitKey(0)
```

# Prewitt Filter

```python
def imgPrewitt(src_img):
    kernelx=np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
    kernely=np.array([[1,1,1],[0,0,0],[-1,-1,1]])
    img_prewittx=cv2.filter2D(src_img,-1,kernelx)
    img_prewitty=cv2.filter2D(src_img,-1,kernely)

    plt.subplot(2,2,1), plt.imshow(src_img, cmap='gray')
    plt.title('Original'),plt.xticks([]),plt.yticks([])
    plt.subplot(2,2,2),plt.imshow(img_prewittx, cmap='gray')
    plt.title('prewittx'),plt.xticks([]),plt.yticks([])
    plt.subplot(2,2,3),plt.imshow(img_prewitty, cmap='gray')
    plt.title('prewitty'),plt.xticks([]),plt.yticks([])
    plt.subplot(2,2,4),plt.imshow(img_prewittx+img_prewitty, cmap='gray')
    plt.title('prewittxy'),plt.xticks([]),plt.yticks([])

    plt.show()
    
imgPrewitt(src_img)
cv2.waitKey(0)
```
